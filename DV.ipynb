```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# -----------------------------
# Dataset
# -----------------------------
data = {
    "Title": [
        "A Light in the Attic", "Tipping the Velvet", "Soumission",
        "Sharp Objects", "Sapiens", "The Hobbit",
        "Harry Potter", "The Alchemist",
        "Atomic Habits", "Rich Dad Poor Dad"
    ],
    "Price": [51.77, 53.74, 50.10, 47.82, 54.23, 35.99, 42.50, 29.99, 39.99, 24.99],
    "Rating": [3, 1, 1, 4, 5, 4, 5, 3, 5, 4]
}

df = pd.DataFrame(data)

# Style
sns.set_style("whitegrid")

# -----------------------------
# 1. Bar Chart - Book Prices
# -----------------------------
plt.figure(figsize=(10,5))
sns.barplot(x="Title", y="Price", data=df, palette="viridis")
plt.xticks(rotation=45, ha='right')
plt.title("Book Prices Comparison")
plt.tight_layout()
plt.show()

# -----------------------------
# 2. Histogram - Price Distribution
# -----------------------------
plt.figure(figsize=(8,5))
sns.histplot(df["Price"], bins=5, kde=True, color="skyblue")
plt.title("Price Distribution of Books")
plt.show()

# -----------------------------
# 3. Count Plot - Ratings
# -----------------------------
plt.figure(figsize=(6,4))
sns.countplot(x="Rating", data=df, palette="Set2")
plt.title("Rating Distribution")
plt.show()

# -----------------------------
# 4. Scatter Plot - Price vs Rating
# -----------------------------
plt.figure(figsize=(7,5))
sns.scatterplot(x="Rating", y="Price", data=df, s=100, color="red")
plt.title("Price vs Rating Relationship")
plt.show()

# -----------------------------
# 5. Box Plot - Outliers
# -----------------------------
plt.figure(figsize=(6,4))
sns.boxplot(x=df["Price"], color="orange")
plt.title("Price Outlier Detection")
plt.show()

# -----------------------------
# 6. Heatmap - Correlation
# -----------------------------
plt.figure(figsize=(5,4))
sns.heatmap(df[["Price", "Rating"]].corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()

print("DATA VISUALIZATION COMPLETED SUCCESSFULLY!")
```
